```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows',None)
```

```python
data = pd.read_csv('learningSet.csv') 
```

```python
nulls_percent_df = pd.DataFrame(data.isna().sum()*100/len(data)).reset_index()
nulls_percent_df.columns = ['column_name', 'nulls_percentage']
nulls_percent_df
```

```python
columns_above_threshold = nulls_percent_df[nulls_percent_df['nulls_percentage']>50]
```

```python
drop_columns_list = list(columns_above_threshold['column_name'])
```

```python
data = data.drop(drop_columns_list, axis=1)
data.shape
```

```python
data['MAILCODE'] = data['MAILCODE'].apply(lambda x: x.replace(" ", "A"))
```

```python
data = data.apply(lambda x: x.replace(" ", np.NaN))
```

```python
nulls_percent_df_2 = pd.DataFrame(data.isna().sum()*100/len(data)).reset_index()
nulls_percent_df_2.columns = ['column_name', 'nulls_percentage']
nulls_percent_df_2
```

```python
columns_above_threshold_2 = nulls_percent_df_2[nulls_percent_df_2['nulls_percentage']>50]
print(len(columns_above_threshold_2['column_name']))
```

```python
drop_columns_list_2 = list(columns_above_threshold_2['column_name'])
```

```python
drop_columns_list_2.extend(['OSOURCE', 'ZIP'])
```

```python
data = data.drop(drop_columns_list_2, axis=1)
data.shape
```

```python
categorical = data.select_dtypes(object)
categorical.head()
```

```python
for col in categorical.columns:
    display(categorical[col].value_counts(dropna=False))
```

```python
df = pd.DataFrame(data['STATE'].value_counts()).reset_index()

df.columns = ['state', 'count']
df
other_states = list(df[df['count']<2500]['state'])
other_states
def clean_state(x):
    if x in other_states:
        return 'other'
    else:
        return x
    
categorical['STATE'] = list(map(clean_state, categorical['STATE']))
```

```python
drop_list = ['NOEXCH', 'MDMAUD','MDMAUD_R', 'MDMAUD_F','MDMAUD_A']
```

```python
categorical['GENDER'] = categorical['GENDER'].fillna('F')
```

```python
categorical['GENDER']=np.where(categorical['GENDER'].isin(['U', 'J', 'C','A']),'other',categorical['GENDER'])
```

We will work on the column DOMAIN. Let's replace the null values with the category that is represented the most and then we will split the data into two columns DOMAIN_A and DOMAIN_B.

- DOMAIN_A will consist of the first character from the DOMAIN
- DOMAIN_B will consist of the second character from the DOMAIN

Then we will drop the original column DOMAIN

```python
categorical['DOMAIN'] = categorical['DOMAIN'].fillna('R2')
```

```python
categorical['DOMAIN_A'] = list(map(lambda x: x[0], categorical['DOMAIN']))
categorical['DOMAIN_B'] = list(map(lambda x: x[1], categorical['DOMAIN']))
```

```python
categorical.DOMAIN_B.value_counts().index
```

We turn it as integer so it will already be ordinal encoded

```python
categorical.DOMAIN_B = categorical.DOMAIN_B.astype(int)
```

```python
categorical = categorical.drop(columns=['DOMAIN'])
```

We will keep the column RFA_2A. We will delete rest of the columns 

```python
for col_name in categorical.columns:
    if "RFA" in col_name:
        drop_list.append(col_name)  
```

```python
drop_list.remove('RFA_2A')
```

```python
categorical = categorical.drop(columns=drop_list)
```

```python
categorical['CLUSTER'] = categorical['CLUSTER'].fillna('40')
```

```python
# categorical['CLUSTER'] = categorical['CLUSTER'].astype(str)
```

```python
categorical['DATASRCE'] = categorical['DATASRCE'].fillna('0')
```

```python
# categorical['DATASRCE'] = categorical['DATASRCE'].astype(str)
```

```python
categorical['GEOCODE2'] = categorical['GEOCODE2'].fillna('A')
```

```python
for col in categorical.columns:
    display(categorical[col].value_counts(dropna=False))
```

```python
for col in categorical.columns:
    display(categorical[col].value_counts().index)
```

```python
categorical.isna().sum()
```

```python
categorical['HOMEOWNR'] = categorical['HOMEOWNR'].fillna('U')
```

```python
categorical = categorical.drop(columns=['AGEFLAG'])
```

```python
categorical.isna().sum().sum()
```

```python
categorical.to_pickle('categorical.pkl')
```
